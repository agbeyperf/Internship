```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("/content/Dataset.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# Missing Values
missing_values = df.isnull().sum().sort_values(ascending=False)
print(missing_values)
```

```python
df['sex'] = df['sex'].astype(str).replace({"Male": 0, "Female": 1})
df['target'] = df['target'].astype(str).replace({"False": 0, "True": 1})
df
```

```python
# A. Getting rid of rows with NAN values
df_dropped_rows = df.dropna()
df_dropped_rows

# B. Replacing missing values with mean value
df_filled_mean = df.fillna(df.mean(numeric_only=True))
df_filled_mean
```

```python
# A
df_dropped_rows.isna().sum()

# B
df_filled_mean.isna().sum()
```

```python
# Grouping the dataset in dependent and independent variables
# A
x_d = df_dropped_rows.drop(['target'], axis=1)
y_d = df_dropped_rows['target']

# B
x = df_filled_mean.drop(['target'], axis=1)
y = df_filled_mean['target']
```

```python
# A
x_d.isna().sum()

# B
x.isna().sum()
```

```python
# Encoding categorical columns with One Hot
# A
x_encoded_d = pd.get_dummies(x_d)
x_encoded_d

# B
x_encoded = pd.get_dummies(x)
x_encoded
```

```python
# A
x_train_d, x_test_d, y_train_d, y_test_d = train_test_split(x_encoded_d, y_d, test_size=0.2, random_state=42)

# B
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)
```

```python
# Standardization
# A
scaler = StandardScaler()
x_train_scaled_d = scaler.fit_transform(x_train_d)
x_test_scaled_d = scaler.transform(x_test_d)

# B
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# A
pd.DataFrame(x_train_scaled_d)

# B
pd.DataFrame(x_train_scaled)
```

```python
# A
pd.DataFrame(x_test_scaled_d)

# B
pd.DataFrame(x_test_scaled)
```

```python
# A
x_train_scaled_d.shape

# B
x_train_scaled.shape
```

```python
# A
# 1. K-Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(x_train_scaled_d, y_train_d)
y_pred_d = knn.predict(x_test_scaled_d)

# Calculating the accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test_d, knn.predict(x_test_scaled_d))
print("Accuracy for Dropped_row DataFrame:", accuracy)

# Model Evaluation (Report and Matrix)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test_d, knn.predict(x_test_scaled_d)))
```

```python
# B
# 1. K-Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(x_train_scaled, y_train)
y_pred = knn.predict(x_test_scaled)

# Calculating the accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for Mean DataFrame:", accuracy)

# Model Evaluation (Report and Matrix)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, knn.predict(x_test_scaled)))
```
